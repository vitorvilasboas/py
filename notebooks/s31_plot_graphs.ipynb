```python
x = np.linspace(0, 2*np.pi, 1000)
f = np.sin(x)
g = np.cos(x)
```

```python
plt.plot(x, f)
plt.plot(x, g)
```

```python
plt.figure(figsize=(10,7))
plt.plot(x, f, linewidth=5, label='$\sin(x)$')
plt.plot(x, g, linewidth=5, label='$\cos(x)$')
plt.xlim()
plt.ylim()
plt.xticks([0,2,4,6])
plt.yticks([-1,-0.5,0,0.5,1])
plt.xlabel('eixo x')
plt.ylabel('eixo y')
plt.title('Seno e Cosseno')
plt.grid(axis='y')
plt.legend()
plt.savefig('datasets/figura1.png', format='png', dpi=300, transparent=True, bbox_inches='tight')
```

```python
plt.figure(figsize=(10,7))
plt.plot(x, f, linewidth=5, label='$\sin(x)$', ls='--') # --  -.  :
plt.plot(x, g, linewidth=5, label='$\sin(x)$')
plt.xlim()
plt.ylim()
plt.xticks([0, np.pi/2, np.pi, 3/2*np.pi, 2*np.pi],
           ['0',r'$\dfrac{1}{2}\pi$',r'$\pi$',r'$\dfrac{3}{2}\pi$',r'$2\pi$'])
plt.yticks([-1,-0.5,0,0.5,1])
plt.xlabel('eixo x')
plt.ylabel('eixo y')
plt.title('Gráfico de Curvas')
plt.grid(axis='y')
plt.legend()
plt.savefig('figura1.png', format='png', dpi=300, transparent=True, bbox_inches='tight')
```
