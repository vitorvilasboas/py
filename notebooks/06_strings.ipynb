* É possível dividir os termos em uma string utilizando os índices como referência
* Define-se intervalos de índices (range) passados entre colchetes
* **Atenção** o índice que encerra o intervalo não é considerado

```python
cadeia = "Aula Sobre Strings em Python" # 28 indices N-1 = 27
```

```python
s1 = cadeia[0:8] # o ':' representa 'até' incluí elementos nos índices de 0 até 7, uma vez que 8 serve apenas como flag (saída do intervalo)
s2 = cadeia[0:9]
s3 = cadeia[3:15]
```

```python
print(f'trecho extraído 0-8: {s1}')
print(f'trecho extraído 0-9: {s2}')
print(f'trecho extraído 3-15: {s3}')
```

```python
s4 = cadeia[:8] # captura todos os literais nos índices até a posição 8 (nesse caso desde o início da string)
```

```python
s5 = cadeia[8:] # captura todos os valores da posição 8 até o final da string
```

```python
print(s4)
print(s5)
```

```python
cadeia[-1] # retorna o valor na última posição de cadeia
```

```python
junta_trechos = s4 + s5 # em strings o operador soma (+) realiza uma concatenação (junção) entre duas strings
```

```python
print('Concatenação: ', junta_trechos)
print('Cadeia: ', cadeia)
```

```python
cadeia[2:20:3] # fatiamento da string cadeia do índice 2 até a 20 com salto de 3 entre os indices nesse intervalo
```

```python
cadeia[2::2] # fatiamento da string cadeia do índice 2 até a 20 com salto de 3
```

**Exemplo**: Leia o nome completo de uma pessoa e imprima:
* o nome com todas as letras maiúsculas
* as partes do nome separadas (primeiro nome, segundo nome, terceiro nome...) 
* quantas letras 'o' e 's' têm no nome completo
* quantos elementos têm o primeiro nome




```python
nome = input('Escreva seu nome completo: ')
mai = nome.upper()
sep = nome.split()
os = nome.count('o')
ss = nome.count('s')
elements = len(sep[0])
print(f"Seu nome completo em maiúsculo: {mai}")
print(f"As partes do seu nome separadas são, respectivamente: {sep}")
print(f"Ocorrências da letra o: {os}")
print(f"Ocorrências da letra s: {ss}")
print(f"Número de elementos no primeiro nome: {elements}")
```

**Exemplo**: Solicite uma senha do usuário e mostre:
* se a senha começa com a palavra "eu"
* se a senha possui a palavra "senha"
* se a senha possui letras, números e caracteres especiais

```python
senha = input("Digite uma senha: ")
eu = 'eu' in senha[0:2]
sen = 'senha' in senha
letra = senha.isalpha()
numero = senha.isnumeric()
lnc = senha.isalnum()
print(f"A senha começa com a palavra 'eu': {eu}")
print(f"A senha possui a palavra 'senha': {sen}")
print(f"A senha possui apenas letras: {letra}")
print(f"A senha possui apenas números: {numero}")
print(f"A senha não possui caracteres especiais: {lnc}")
```

**Exemplo**: Peça para o usuário informar uma frase qualquer e depois uma palavra qualquer e mostre:
* o comprimento da palavra informada
* o caractere na frase cujo índice é igual ao comprimento da palavra
* em que posição a letra "e" aparece na frase pela primeira vez
* a frase com sua primeira palavra substituída pela palavra fornecida pelo usuário

```python
frase = input("Escreva uma frase qualquer: ")
palavra = input("Escreva uma palavra qualquer: ")
compp = len(palavra)
relac = frase[compp]
e = frase.find("e")
split = frase.split()
subtt = frase.replace(split[0], palavra)
print(f"O comprimento da palavra escrita é: {compp}")
print(f"O caractere que corresponde ao numero do comprimento da palavra é: {relac}")
print(f"A letra 'e' aparece pela primeira vez na frase no índice {e}")
print(subtt)
```

**Exemplo**: Leia uma senha do usuário e retorne o quadrado do número de ocorrências do caractere "a" caso ele apareça pelo menos 3 vezes, caso não, retorne o cubo do número de ocorrências.

```python
frase = input("Entre com uma senha qualquer: ")

n = frase.count("a")

if n >= 3:
  r = n**2
  print(f"O caractere 'a' aparece {n} vezes na frase informada.")
  print(f"resultado: {r}")
else:
  r = n**3
  print(f"O caractere 'a' aparece {n} vezes na frase informada.")
  print(f"resultado: {r}")
```

**Exemplo**: Crie uma função básica que retorna True se a palavra 'dog' estiver contida na string de entrada. Não se preocupe com os casos de extremos, como uma pontuação que está sendo anexada à palavra cão, mas que seja senível à caixa.

```python
def encontreCachorro(st):
    return 'cachorro' in st.lower().split()
```

```python
encontreCachorro('Existe um cachorro aí?')
```

**Exemplo**: Crie uma função que conta o número de vezes que a palavra "dog" ocorre em uma string. Novamente ignore os casos extremos.

```python
def contaCachorro(st):
    count = 0
    for word in st.lower().split():
        if word == 'cachorro':
            count += 1
    return count
```

```python
contaCachorro('Esse cachorro é mais rápido que o outro cachorro.')
```

**Exemplo**: Use expressões lambda e a função filter () para filtrar as palavras de uma lista que não começa com a letra 's'. Por exemplo:

    seq = ['sopa','cachorro','salada','gato','ótimo']

Deveria ser filtrado para:

    ['sopa','salada']

```python
seq = ['sopa','cachorro','salada','gato','ótimo']
```

```python
print(list(filter(lambda palavra: palavra[0]=='s', seq)))
```
