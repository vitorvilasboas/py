[Python](http://www.python.org/) é uma linguagem de programação dinâmica, de altíssimo nível, orientada a objeto, interpretada, funcional e interativa. Destaca-se como uma linguagem simples e versátil, que pode ser usada no desenvolvimento de diversos tipos de aplicações: web, desktop, mobile, IoT (Internet of Things), sistemas inteligentes, dentre outras.  

## Características da linguagem

O Python possui uma sintaxe clara e concisa, que favorece a legibilidade do código fonte e agiliza o aprendizado. O uso do Python é frequentemente associado à ganhos significativos de produtividade e à produção de programas de alta qualidade e de fácil manutenção. A linguagem inclui diversas estruturas de alto nível (listas, dicionários, arrays e outras) e uma vasta coleção de módulos nativos, prontos para uso, além de pacotes de terceiros que podem ser adicionados. Também possui recursos encontrados em outras linguagens modernas, tais como: geradores, introspecção, persistência, metaclasses e unidades de teste. 

Por ser multiparadigma, a linguagem Python suporta programação modular e funcional, além da orientação a objetos. Mesmo os tipos de dados mais básicos no Python são objetos. Outra característica importante relacionada à orientação a objetos é que o acesso há classes no Python é menos restrito se comparado à outras linguagens como C++ e Java, por exemplo.

Python é *open source* (de código aberto), com licença compatível com a General Public License (GPL), porém menos restritiva, permitindo que o Python seja inclusive incorporado em produtos proprietários. A linguagem começou a ser desenvolvida no final dos anos 80 por Guido van Rossum no Instituto Nacional de Pesquisa para Matemática e Ciência da Computação da Holanda (CWI), sendo lançada oficialmente em meados de 1991. Originalmente o Python foi concebido a partir de outra linguagem existente na época, chamada ABC, e era voltada para um perfil específico de usuários formado por físicos e engenheiros.

Atualmente, sua especificação é mantida pela [Python Software Foundation](https://www.python.org/psf/) (PSF), uma fundação sem fins lucrativos que reune notórios programadores e entusiastas da comunidade python ao redor do mundo. A implementação oficial do Python é escrita em C, e por isso, é também conhecida como CPython. 

Novas versões da linguagem são validadas e lançadas periódicamente, trazendo consigo correções, alterações na sintaxe, recomendações de implementação, novos módulos utilitários, dentre outras melhorias. A versão estável mais recente está disponível para download neste [link](http://www.python.org/download/).

É importante observar que o Python não se resume às regras sintáticas, léxicas e semânticas, a linguagem também engloba [documentação oficial](https://docs.python.org/), a Python Virtual Machine (PVM), centenas de módulos nativos, bem como outros [pacotes e ferramentas](https://pypi.org/), dentre as quais está o depurador integrado [IDLE](https://docs.python.org/3/library/idle.html) - Ambiente Integrado de Desenvolvimento e Aprendizagem (do inglês, *Integrated Development and Learning Environment*).

Em termos gerais, a linguagem apresenta muitas similaridades com outras linguagens dinâmicas, como Perl e Ruby. Além disso, é possível integrar o Python a outras linguagens, como a Linguagem C e Fortran, por exemplo, quando atua como uma linguagem de script que permite automatizar tarefas e adicionar novas funcionalidades à softwares já consagrados. BrOffice.org, PostgreSQL, Blender, GIMP e Inkscape são exemplos de sistemas desenvolvidos em outras linguagem mas que usam scripts python em seu escopo. A linguagem é usada em aplicações de grandes empresas de tecnologia como Google, Yahoo, Microsoft, Nokia e Disney.

## Por que Python?

Em suma, dentre as várias vantagens de criar programas em Python, destacam-se:
* **Multiplataforma**: pode ser instalada em qualquer plataforma computacional, desde PDAs até supercomputadores com processamento paralelo, passando por todas as plataformas de computação pessoal.
* **Portabilidade**: Aplicativos desenvolvidos em Python podem ser facilmente distribuídos para várias plataformas diferentes daquela
em que foi desenvolvido, mesmo que estas não possuam o Python instalado.
* **Software livre**: O Python é software livre, não impondo qualquer limitação à distribuição gratuita ou venda de programas.
* **Extensibilidade**: O Python pode ser extendido através de módulos, escritos em Python ou rotinas escritas em outras linguagens, tais como C ou Fortran.
* **Orientação a objeto**: Tudo em Python é um objeto. Funções, variáveis de todos os tipos e até módulos (programas escritos em Python) são objetos.
* **Tipagem automática**: Python é uma linguagem dinamicamente tipada. Não é necessário tipar as variáveis para usá-las. O tipo de uma variável (string, inteiro, float, etc.) é determinado durante a execução do código.
* **Código legível**: por utilizar uma sintaxe simplificada e forçar a divisão de blocos de código por meio de indentação, o python torna-se bastante legível, mesmo para pessoas que não estejam familiarizadas com o programa.
* **Flexibilidade**: já conta com módulos para diversas aplicações, científicas ou não, incluindo módulos para interação com os protocolos mais comuns da Internet (FTP, HTTP, XMLRPC, etc.). A maior parte destes módulos já faz parte da distribuição básica do Python.
* **Operação com arquivos**: A manipulação de arquivos, tais como a leitura e escrita de dados em arquivos texto e binário, é muito simplificada no Python, facilitando a tarefa de muitos pesquisadores ao acessar dados em diversos formatos.
* **Uso interativo**: O Python pode ser utilizado interativamente ou invocado para a execução de scripts completos. O uso interativo permite “experimentar” comandos antes de incluí-los em programas mais complexos, ou usar o Python simplesmente como uma calculadora.

## O Interpretador Python

A linguagem Python é interpretada através de *bytecode* pela máquina virtual Python - PVM, tornando o código portável para multiplas plataformas. A PVM é definida por um programa escrito nas linguagens C e C++, o qual é o responsável por interpretar e executar os códigos-fontes escritos em Python. Com isso é possível compilar aplicações em uma plataforma e executar em outras plataformas que contenham Máquina Virtual do Python instalada (multiplataforma).

## Tipos de dados em Python 


**Dado** é um conjunto de valores "pura e simplesmente". 

Um dado pode ser um número, uma palavra, uma imagem, enfim, algo coletado, medido ou criado. 

Dados não precisam transmitir uma idéia! Por exemplo, o número 7, por si só, não transmite uma relação com nenhum contexto, é apenas o número 7. No entanto, ao definir o contexto, como dizer "Rua 7", o dado passa a transmitir uma ideia e passamos a ter uma **informação**.

Na computação, todo dado é representado por uma sequência de bits (ex: 0111).

O processamento e o armazenamento de um dado pelo computador depende de sua classificação (tipo do dado).

Tipos primitivos:
* **int** define um valor inteiro...
* **float** define um valor real (valores decimais)
* **bool** define um valor boobleano (Verdadeiro ou Falso)
* **str** define uma string (cadeia de caracteres - representa textos)

A **TIPAGEM** de variáveis no Python é **DINÂMICA**

Entrada e Saída **formatadas**

````python
nome, idade = 'Vitor', 35
````

````python
print("O nome informado é {} e a idade informada é {} anos!".format(nome, idade)) # função de saída com método FORMAT
````

````python
print("O nome informado é %s e a idade informada é %d anos!" %(nome, idade)) # função de saída baseada no endereço de memória (%s=string  %d=inteiro/decimal  %f=float)
````

````python
x, y = 4, 3
````

````python
print('A soma é %d, o produto é %d e a divisão é %.2f' % (x+y, x*y, x/y)) # .2f = duas casas após a vírgula 
````

````python
print('A soma é {}, o produto é {} e a divisão é {:.2f}'.format(x+y, x*y, x/y), end='\n\n\n')  # 'end' altera o passo após imprimir (aqui 3 quebras de linha)
````

````python
print('='*20) # operadores aritméticos na função de saída padrão
````

````python
print("Meu nome é {:<20}".format(nome)) # saída FORMAT espaçada - 20 espaços após o argumento
````

````python
print("Meu nome é {:>20}".format(nome)) # saída FORMAT espaçada - 20 espaços antes do argumento
````

````python
print("Meu nome é {:^20}".format(nome)) # saída FORMAT espaçada - argumento centralizado em 20 espaços (10 antes e 10 depois)
````

**Exemplo:** Leia o nome do colaborador de uma FINTEC e a quantidade de clientes indicados por ele no último mês. Sabendo que a empresa paga a ele um salário fixo de R\$5000,00 por mês mais um bônus de R\$50,00 a cada novo cliente indicado, calcule e mostre o valor total do salário a que o colaborador tem direito a receber no mês (fixo + bônus).

````python
fixo = 5000
bonus = 50
nome = input("Informe o nome do colaborador: ")
qtd = int(input("Informe o número de clientes atendidos: "))
salario = fixo + (qtd * bonus)
print(f"O salário a receber do colaborador {nome} é: R${salario}")
````

**Exemplo**: Crie um script Python capaz de mostrar o valor 3.1415926535 com 1, 2, 3 e 5 casas decimais.

````python
pi = 3.1415926535

print("Valor de PI com UMA casa decimal: %.1f" % pi)
print("Valor de PI com DUAS casas decimais: %.2f" % pi)
print("Valor de PI com TRÊS casas decimais: {:.3f}".format(pi))
print("Valor de PI com CINCO casas decimais: %.5f" % pi)


print(f"Valor de PI com CINCO casas decimais: {round(pi,5)}")
````

**Exemplo**. Leia um salario (real) e um percentual (inteiro) de aumento. Informe o salario com aumento em duas casas decimais.

````python
salario = float(input("Informe o salario (R$): "))
aumento = float(input("Informe o valor do aumeto (%): "))
calculo = salario * aumento/100
salario += calculo # salario = salario + calculo
print(f"Novo salário: R${round(salario,2)}")
````
