#### Condicionais:
* São usadas para controle de fluxo (desvios no código)
* Os desvios ocorrem caso determinada condição seja satisfeita
* Podem ser simples ou compostas
* Podem ser aninhadas (uma "dentro" da outra)
* Importante: os blocos de código dentro das estruturas condicionais são identificados pela **indentação** (tecla *Tab*)
* Utiliza-se com frequência expressões/operadores relacionais (==, !=, >, <, >=, <=, %)
* Também adota-se expressões lógicas/booleanas (and, or, not, in, not in, is, is not)

P.S.: **No python o uso da indentação é obrigatório** para delimitar blocos de código em condicionais, repetições, funções, classes, etc.... O início dos blocos se dá pelo operador **:** (dois pontos). Diferentemente do Java, por exemplo, em que a indentação não é obrigatória (mas recomendável) e os blocos são escritos entre chaves (delimitadores *{* e *}*). 

Embora o raciocínio lógico seja o componente mais importante na programação, é importante conhecer as diferenças de sintaxes ao longo do percurso. Isso adiciona experiência.

#### Condicionais simples, compostas e aninhadas

**Exemplo**. Solicite um número do usuário e responda se ele é igual à 5. 

```
num = input("Informe um número: ")
num_int = int(num)  # converte de str para inteiro
print("O valor informado é igual a  5?\n")
```

```
resp = (num_int == 5)
print(resp)
```

#### Condicional simples (se)

```
if num_int == 5: # condicional: se <num> é igual a 5 então...
  print("sim.\n")   # instrução de execução condicionada verdadeira ... observe a indentação!

print("... ok, obrigado.")
```

#### Condicional composta (se.. senão)

```
if num_int == 5:
  print("sim.\n")  # bloco verdadeiro (True)
  print(f"Pode haver várias instruções num bloco!\n")
else:  ## senão
  print("não.\n")  # bloco falso (False)

print("... ok, obrigado.")
```

#### Condicionais aninhadas (se.. senão se.. senão)

```
if num_int is 5:
  print("sim.\n")
elif num_int is 10: ## senão.. se
    print(f"não! O valor é {num_int}.\n")
elif num_int is 20:
  print(f"não! E também não é 10! O valor é {num_int}.\n")
else:
  print(f"não! E também não é 10 e nem 20! O valor é {num_int}.\n")

print("Obrigado.")
```

```
gratuito = False
preço = 300.00

if gratuito:
  print("Legal, quero aprender agora!")
elif preço <= 100.00:
  print("Quero comprar!")
elif preço < 250:
  print("Deixa eu pensar uns 5 minutinhos")
else:
  print("Quem sabe no futuro, agora não posso!")
```

```
print("O valor informa é maior que 5?\n")

if num_int > 5: print("sim.\n")   # em uma linha só! característica funcional do python
else: print("não.\n")

print("Ok, obrigado.")
```

#### Condicionais com múltiplos testes

**Exemplo**. Solicite um número do usuário e responda se ele está entre 0 e 1. 

```
real = float(input("Informe um número: ")) # solicita entrada do usuário (texto - string - str) e já converte para o tipo numérico real (float)

if (real >= 0) and (real <= 1): # as instruções dentro do bloco só são executadas se as duas condições forem satisfeitas
  print(f"O número informado, {real}, está contido no intervalo entre 0 e 1.\n")
else:
  print(f"O número informado, {real}, está FORA do intervalo entre 0 e 1.\n")
print("Programa encerrado!")
```

```
if (real >= 0) and (real <= 1): 
  print(f"O número informado, {real}, está contido no intervalo entre 0 e 1.\n")
elif (real > 1):
  print(f"O número informado, {real}, é positivo maior que 1.\n")
else:
  print(f"O número informado, {real}, é negativo.\n")
print("Programa encerrado!")
```

**Exemplo**. Faça um programa que leia duas notas, calcule a média aritmética entre elas e diga se o usuário foi: 
* Aprovado (media maior que 6)
* Vai para conselho (media entre 4 e 5,99) 
* Reprovado (média menor que 4)

```
n1 = float(input("Informe a primeira nota: "))
n2 = float(input("Informe a segunda nota: "))
media = (n1+n2)/2
print(f"Média: {media}")
if media >= 6:
  print("Aprovado")
elif (media>=4) and (media < 6):
  print("Vai para conselho")
else:
  print("Reprovado")
```

```
media = ( float(input("Informe a primeira nota: ")) + float(input("Informe a segunda nota: ")) ) / 2
print("Média: {}\nAprovado".format(media) if (media >= 6) else "Média: {}\nVai para conselho".format(media) if ((media>=4) and (media < 6)) else "Média: {}\nReprovado".format(media))
```
