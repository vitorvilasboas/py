```python
# somar duas listas multidimensionais (matrizes)
x = [[2,3,3],[4,5,6],[4,1,2]]
y = [[7,1,5],[2,8,3],[9,2,4]]
s = [[],[],[]]
for i in range(3):
  for j in range(3):
    s[i].append(x[i][j] + y[i][j])
print(s)
```

```python
# somar todos os elementos de uma lista multidimensional (matriz)
soma = 0
for i in range(3):
  for j in range(3):
    soma += x[i][j] 
print(f"Soma: {soma}")
```

```python
# somar os elementos de cada uma das linhas de uma lista multidimensional (matriz)
for i in range(3):
  soma = 0
  for j in range(3):
    soma += x[i][j] 
  print(f"Soma linha {i}: {soma}")
```

```python
# apresentar o maior valor e sua posição em uma lista multidimensional (matriz)
maior = x[0][0]
for i in range(3):
  for j in range(3):
    if x[i][j] > maior: maior = x[i][j]

pos_i, pos_j = 0, 0
for i in range(3):
  for j in range(3):
    if x[i][j] == maior: pos_i, pos_j = i, j

print(f"Maior: {maior} .. posição [{pos_i}][{pos_j}].")
```

```python
# apresentar os elementos da diagonal principal de uma lista multidimensional (matriz)
for i in range(3):
  for j in range(3):
    if i == j: print(" "*(i*3) + f"{x[i][j]}")
```

```python
# apresentar os elementos acima e abaixo da diagonal principal em uma lista multidimensional (matriz)
for i in range(3):
  for j in range(3):
    if i != j: print(" "*(j*3) + f"{x[i][j]}", end="")
  print("\n")
```

**Exemplo**: Ler um inteiro (N) e imprimir os N primeiros números primos

```python
# Ler um inteiro (N) e imprimir os N primeiros números primos

N = int(input("Informe um número inteiro: "))

primos = []

num = 1
while(len(primos)< N ):
  cont = 0
  for i in range(1, num+1):
    if num % i == 0:
      cont += 1
  if cont <= 2:
    primos.append(num)
  num += 1

print(f"Lista dos {N} primeiros números primos: {primos}")
```

**Exemplo**: Leia o salário de cinco funcionários e calcule o desconto do imposto de seguridade social de cada um deles sabendo que para salários superiores a R\$5000 o desconto é de 22% e para salários até R\$5000 o desconto é de 10%. Mostre o salário mais alto após o desconto.

```python
salarios = []
for i in range(0, 5):
  salario = float(input(f"Entre com o salário do {i+1}o funcionário: R$"))
  
  if salario > 5000:
    salario -= (salario * 0.22)
  else:
    salario -= (salario * 0.10)

  salarios.append(salario)
  print(f"==> Salário ajustado: R${salario:.2f}\n")

maior = salarios[0]
for salario in salarios:
  if salario > maior:
    maior = salario
print(f"==> Maior salário ajustado: R${maior:.2f}")
```

**Exemplo:** Crie uma função que receba uma lista de números fornecidos pelo usuário e retorne uma nova lista cujos elementos são o dobro dos elementos na lista inicial. 

```python
def calculoDobro(lista1):
  lista2 = list()
  for x in lista1:
    lista2.append(x**2)
  return lista2

# percorrendo lista
lista = list()
while True:
  num = int(input("Informe um número inteiro. Para sair digite -99: ")) # flag (bandeira) para saida do laço de repetição é -99
  if(num != -99):
    lista.append(num)
  else: # se o número informado for igual à -99
    print("Ok, você saiu do laço! ")
    break # para a execução do laço no momento em que o flag foi informado

print(f"Lista inicial: {lista}")

quadrados = calculoDobro(lista)

print(f"Lista resultante: {quadrados}")
```

**Exemplo**: Leia uma quantidade indeterminada de números reais. A cada número lido o programa deve perguntar se o usuário deseja informar um novo número, ler e armazenar o próximo número enquanto a resposta for positiva. Ao final, assim que o usuário desejar encerrar a entrada de novos números, mostre:
* A quantidade de números lidos;
* A soma e a média dos números lidos;
* A quantidade de números maiores ou iguais à média de todos os números lidos;
* Quantos pertencem ao intervalo entre 10 e 50;
* O maior e o menor número
* Todos os números na ordem em que foram informados, um ao lado do outro
* Todos os números na ordem inversa à que foram informados, um abaixo do outro



```python
# 1o passo: solicitar e armazenar os números em uma lista

lista = [] # crio uma lista vazia
while(True):
  num = float(input("Informe um número real: "))
  lista.append(num)  # armazeno o número informado dentro da lista
  if input("Quer informar outro número? [S]im ou [N]ão: ").upper() == 'N': 
    break

#2o passo: verificar o tamanho da lista == qtd de níumeros dentro dela
qtd = len(lista)
print(f"Foram informados {qtd} números. ")

#3o passo: percorrer a lista, somar todos os números e calcular a média
soma = 0
for n in lista: # percorrer a lista
  soma += n
media = soma/qtd
print(f"A soma dos números informados é {soma} e a média é {media}.")

qtd_maior = 0
qtd_intervalo = 0

for n in lista:
  if n >= media:
    qtd_maior += 1
  if n >= 10 and n <= 50:
    qtd_intervalo += 1
  
print(f"Quantidade de números maiores ou iguais a média é {qtd_maior}.")
print(f"Quantidade de números entre 10 e 50 é {qtd_intervalo}.")

# O maior e o menor número
# Todos os números na ordem em que foram informados, um ao lado do outro
# Todos os números na ordem inversa à que foram informados, um abaixo do outro

menor = lista[0]
maior = lista[0]
for num in lista:
  if num < menor:
    menor = num
  if num > maior: 
      maior = num

print("\n")
print(f"{menor} foi o menor número informado e {maior} o maior.")

print("\n")
print("Lista: ", end="")

for i in range(0, len(lista), 1):
  print(lista[i], end=" ")

print("\n")
print("Lista invertida: ")

for i in range(len(lista) - 1, -1, -1):
  print(lista[i])
```

```python
def solicitar_numeros():
  # Função para solicitar os números e armazená-los numa lista
  lista = [] # lista vazia
  while(True):
    num = float(input("Informe um número real: "))
    lista.append(num)  # armazeno o número informado dentro da lista
    if input("Quer informar outro número? [S]im ou [N]ão: ").upper() == 'N': 
      break
  return lista

def calcular_media(lista):
  # Função percorre a lista, soma todos os números e calcula a média entre eles
  soma = 0
  for n in lista: # percorrer a lista
    soma += n
  media = soma/len(lista)   
  return soma, media

def contar_maior_media(lista, media):
  # Conta a quantidade de números na lista que são maiores ou iguais à média
  cont = 0
  for n in lista:
    if cont >= media:
      cont += 1
  return cont

def contar_nums_intervalo(lista):
  # Conta a quantidade de números na lista que pertencem ao intervalo entre 10 e 50
  cont = 0
  for n in lista:
    if n >= 10 and n <= 50:
      cont += 1
  return cont

def extrair_menor(lista):
  # Retorna o menor número na lista
  menor = lista[0]
  for num in lista:
    if num < menor:
      menor = num
  return menor

def extrair_maior(lista):
  # Retorna o maior número na lista
  maior = lista[0]
  for num in lista:
    if num > maior: 
      maior = num
  return maior

def imprime_lista(lista):
  # Mostra todos os números na lista na ordem em que foram informados, um ao lado do outro
  print("\nLista: ", end="")
  for i in range(0, len(lista), 1):
    print(lista[i], end=" ")

def imprime_lista_invertida(lista):
  # Mostra todos os números na lista na ordem inversa à que foram informados, um abaixo do outro
  print("\nLista invertida: ")
  for i in range(len(lista) - 1, -1, -1):
    print(lista[i])


# if __name__ == '__main__':

numeros = solicitar_numeros()
s, m = calcular_media(numeros)
qtd_maiorm = contar_maior_media(numeros, m)
qtd_intervalo = contar_nums_intervalo(numeros)
menor = extrair_menor(numeros)
maior = extrair_maior(numeros)
 
print(f"Foram informados {len(lista)} números.") # len(lista) é uma função nativa que retorna o tamanho da lista, ou seja, a qtd de elementos nela
print(f"A soma dos números informados é {s} e a média é {m}.")
print(f"Quantidade de números maiores ou iguais a média é {qtd_maiorm}.")
print(f"Quantidade de números entre 10 e 50 é {qtd_intervalo}.")
print(f"{menor} foi o menor número informado e {maior} o maior.")   # max(numeros) e min(numeros)

imprime_lista_invertida(lista)
imprime_lista(lista)
```

**Exemplo**: Simule partidas do jogo "**pedra, papel, tesoura**" entre o usuário e o computador. A cada partida, determine a escolha aleatória do computador, leia a escolha do usuário e exiba o resultado conforme o exemplo:

Usuário: PAPEL

Computador: TESOURA

"TESOURA vence PAPEL. Vitória do Computador!”

O programa deve realizar diversas partidas e encerrar somente quando o usuário desejar.


```python
opcoes = ['pedra','papel','tesoura']
import random, time
cont_user, cont_comput, jogadas = 0, 0, 0
while True:
  print('''Escolha uma das opções: 
  1 - Pedra
  2 - Papel
  3 - Tesoura
  0 - Sair 
  ''')
  usuario = int(input("Opção: "))
  comput = random.choice(opcoes)
  # comput = random.randint(0, 2)

  if usuario != 0:
    
    if usuario not in [1,2,3]:
      print("Opção inválida!")
    else:
      jogadas += 1
      if (usuario == 1 and comput == 'tesoura') or (usuario == 2 and comput == 'pedra') or (usuario == 3 and comput == 'papel'):
        print(f"Computador escolheu {comput}, logo você venceu!\n")
        cont_user += 1
      elif (usuario == 1 and comput == 'papel') or (usuario == 2 and comput == 'tesoura') or (usuario == 3 and comput == 'pedra'):
        print(f"Computador escolheu {comput} e venceu!\n")
        cont_comput += 1
      else:
        print(f"Computador escolheu {comput} e, portanto, houve empate!\n")

  else:
    print("Saindo", end=" ")
    for i in range(0,3):
      time.sleep(1.5)
      print(".", end=" ")
    break

print(f"\nJogadas:: {jogadas}")
print(f"Placar:: computador = {cont_comput}  || você = {cont_user}")
```

**Exemplo**: Leia o nome, o sexo e a idade de sete pessoas e mostre o nome e a idade da pessoa mais velha e da pessoa mais jovem. Mostre também a quantidade de homens que atingiram a maioridade.

```python
pessoas = []
for i in range(1, 8, 1):
  nome = input(f"Informe o nome da {i}ª pessoa: ")
  sexo = input(f"Informe o sexo da {i}ª pessoa [M/F]: ").upper()
  idade = int(input(f"Informe a idade da {i}ª pessoa: "))
  p = {'nome':nome, 'sexo':sexo, 'idade':idade}
  pessoas.append(p)
  print("\n")

# print(pessoas)

menor = pessoas[0]
maior = pessoas[0]

cont_maioridade = 0

for pes in pessoas:
  if pes['idade'] < menor['idade']:
    menor = pes

  if pes['idade'] > maior['idade']:
    maior = pes

  if pes['sexo']=='M' and pes['idade']>=18:
    cont_maioridade += 1

print(f"{maior['nome']}, com {maior['idade']} anos de idade, é a pessoa mais velha dentre as informadas.")
print(f"{menor['nome']}, com {menor['idade']} anos de idade, é a pessoa mais jovem.")
print(f"{cont_maioridade} homens atingiram a maioridade (18 anos).")
```

**Exemplo**: Leia o nome, o cargo, a idade, o sexo e o salário de vários funcionários de uma empresa. A cada funcionário cadastrado, o programa deverá perguntar se o usuário quer ou não continuar. No final, mostre:
* quantos funcionários com salário acima de R\$2000,00 foram cadastrados.
* os nomes dos funcionários menores de 18 anos.
* qual a média de salário dos funcionários homens.
* qual a média de salário das funcionárias mulheres com mais de 30 anos de idade.



```python
funcionarios = []
while(True):
  op = int(input(f'''
  ==== Cadastro de funcoinários ====
  [1] Cadastrar novo funcionário
  [2] Sair e gerar relatório
  {'='*34}
  opção: '''))
  print("\n")
  if op == 1:
    print('=== Novo Funcionário ===')
    nome = str(input("Nome: "))
    cargo = str(input("Cargo: "))
    idade = int(input("Idade: "))
    sexo = str(input("Sexo [M/F]: ")).upper()
    salario = float(input("Salario: R$"))
    func = {'nome':nome, 'cargo':cargo, 'idade':idade, 'sexo':sexo, 'salario':salario}
    funcionarios.append(func)
  elif op == 2: 
    print("Saindo...")
    break
  else: print('OPÇÃO INVÁLIDA!')

print("\n")
print("=== Relatório ===")

if funcionarios == []: print("Nenhum funcionário cadastrado.")
else: 
  # quantos funcionários com salário acima de R$2000,00 foram cadastrados.
  cont = 0
  for f in funcionarios:
    if f['salario'] > 2000: cont += 1
  print(f" {cont} funcionários têm salário acima de R$2000,00.")

  # nomes dos funcionários menores de 18 anos.
  print(f" {cont} Funcionários menores de 18 anos:")
  for f in funcionarios:
    if f['idade'] < 18: print(f['nome'])

  # média de salário dos funcionários homens e 
  # média de salário das funcionárias mulheres com mais de 30 anos de idade.
  soma_sal_h, cont_h = 0, 0
  soma_sal_m30, cont_m30 = 0, 0
  for f in funcionarios:
    if f['sexo'] == 'M': 
      soma_sal_h += f['salario']
      cont_h += 1
    else:
      if f['salario'] > 30:
        soma_sal_m30 += f['salario']
        cont_m30 += 1

  print(f" Média de salário entre os funcionários do sexo masculno: R${(soma_sal_h/cont_h):2.f}")
  print(f" Média de salário entre os funcionárias mulheres maiores que 30 anos: R${(soma_sal_m30/cont_m30):2.f}")
```

**Exemplo**: Sistema de Recomendação de Cursos. Projete e implemente um sistema de recimendação, a partir de uma lista de cursos, cada qual com titulo, duração em horas, categoria e gratuidade. Após estabelecer uma lista de dicionários de cursos, o sistema deve consultar as preferências do usuário como: categoria de interesse, tempo mínimo de duração e Gratuidade, construir seu perfil e recomendar os cursos, no dicionário, que atendam a esse perfil. Abaixo, um exemplo de saída do sistema:

Olá, **Fulano**, com base no seu perfil, achamos que você iria gostar dos seguintes cursos:

PROGRAMAÇÃO:
- Programação Python (90 horas)
- Desenvolvimento de Plugins para WordPress (160 horas)

IA:
- Machine Learning com Python (100 horas)

```python
nome = input("Como gostaria de ser chamado(a): ")

cursos = [{'titulo': 'Programação Python', 'duracao': 90, 'categoria': 'Programação', 'gratuidade': True},
          {'titulo': 'Machine Learning com Python', 'duracao': 100, 'categoria': 'IA', 'gratuidade': False},
          {'titulo': 'Desenvolvedor Web Full Stack', 'duracao': 160, 'categoria': 'Programação', 'gratuidade': False},
          {'titulo': 'Introdução ao Google Ads', 'duracao': 25, 'categoria': 'Marketing digital', 'gratuidade': True},
          {'titulo': 'E-commerce e Vendas Online', 'duracao': 40, 'categoria': 'Vendas', 'gratuidade': False},
          {'titulo': 'Canva para Empreendedores', 'duracao': 60, 'categoria': 'Marketing digital', 'gratuidade': True},
          {'titulo': 'Desenvolvimento de Plugins para WordPress', 'duracao': 70, 'categoria': 'Programação', 'gratuidade': False}]

lista_categorias = list(set([ cursos[i]['categoria'] for i, curso in enumerate(cursos) ]))

print('\nCategorias de Cursos: ')
for i,cat in enumerate(lista_categorias): print(f"{i}. {cat}")

id_categorias = [ int(n.strip()) for n in input("\nInforme o ID das categorias de seu interesse (separe com vírgulas): ").split(',') ]
tempo = float(input("Quanto tempo gostaria de estudar (duração mínima em horas)? "))
# gratuidade = input("Preferência por Cursos Gratuitos [S/N]? ")
pagos = False if input("Incluir Cursos Pagos [S/N]? ").upper() == "N" else True

categorias_selecionadas = [ lista_categorias[id] for id in id_categorias ]

print(f"\nNome: {nome}")
print(f"Categorias:", end="")
for cat in categorias_selecionadas: print(f" {cat},", end="")
print(f"\nDuração mínima: {tempo} horas")
print(f"Incluir Pagos: {'SIM' if pagos else 'NÃO'}")

print(f"\n{nome}, com base no seu perfil, achamos que você iria gostar dos seguintes cursos: ")

for cat in categorias_selecionadas:
  print(f"\n{cat}:")
  for curso in cursos:
    if (curso['gratuidade'] or pagos) and (cat.capitalize() == curso['categoria'] and tempo <= curso['duracao']):
      print(f"– {curso['titulo']} ({curso['duracao']} horas)")
```

### Sets

Conjuntos - coleção não ordenada e não indexada

```python
conjunto = {"apple", "banana", "cherry"}  # define um conjunto com 3 itens/elementos
print(conjunto)
```

```python
conjunto.add("orange")  # adiciona um elemento no conjunto
print(conjunto)
```

```python
conjunto.remove("banana")   # remove um elemento específico do conjunto
print(conjunto)
```

```python
conjunto.discard("banana")  # remove um elemento específico do conjunto
print(conjunto)
```

```python
x = conjunto.pop()  # retira o 1º item do conjunto (exclusão pela esquerda - fila)
print(x)
```

```python
conjunto.update(["orange", "mango", "grapes"])   # atualiza o conjunto
print(conjunto)
```

```python
conjunto.clear()   # limpa o conjunto (exclui os elementos)
print(conjunto)
```

```python
del conjunto   # exclui o conjunto definitivamente
```
